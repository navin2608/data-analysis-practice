```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("Z:\climate-change-earth-surface-temperature-data\GlobalTemperatures.csv")
df.head(4)
```

```python
import matplotlib.pyplot as plt

plt.style.use('ggplot')
```

```python
df.tail(10)
```

```python
df=df.iloc[:,:2]
df.head()
```

```python
df.describe()
```

```python
plt.figure(figsize=(15,5))
plt.plot(df['LandAverageTemperature'])
plt.show()
```

```python
print(type(df['dt'][0]))
```

```python
times=pd.DatetimeIndex(df['dt'])
grouped=df.groupby([times.year]).mean()
```

```python
grouped.tail()
```

```python
plt.figure(figsize=(20,5))
```

```python
plt.plot(grouped['LandAverageTemperature'])
plt.title('yearly average land temperature')
plt.show()
```

```python
a=df.groupby([times.year]).mean()
a
```

```python
df[np.isnan(df["LandAverageTemperature"])]
```

```python
df['LandAverageTemperature']=df['LandAverageTemperature'].fillna(method='ffill')
```

```python
grouped=df.groupby([times.year]).mean()
plt.figure(figsize=(15,5))
plt.plot(grouped['LandAverageTemperature'])
plt.show()
```

modelling dataset# 

```python
from sklearn.linear_model import LinearRegression as linreg
```

```python
x=grouped.index.values.reshape(-1,1)
y=grouped['LandAverageTemperature'].values
```

```python
grouped
```

```python
reg=linreg()
reg.fit(x,y)
y_preds=reg.predict(x)
print('accuracy'+str(reg.score(x,y)))
```

```python
plt.figure(figsize=(15,5))
plt.title("linear regression")
plt.scatter(x=x,y=y_preds)
plt.scatter(x=x,y=y,c='r')
```
